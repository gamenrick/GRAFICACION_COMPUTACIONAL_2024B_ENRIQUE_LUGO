# Empleando modelo Lotka - Volterra

## ¿Que es el modelo?

### El modelo de lotka volterra es una forma de relacionar el comportamiento de dos especies diferentes en un entorno controlado y la forma en la que el comportamiento de estos los afecta mutuamente

### implementación del modelo en python

```python
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys
```

### Se cuenta con el siguiente código proporcionado en la clase de graficación computacional. Este codigo fue ligeramente modificado para implementarl el metodo principal en forma de funcion. El objetivo de esta práctica es entender dicho código proporcionado y jugar con las variables después de haber analizado dicho código y entenderlo completamente. 

```python
def funcionLotka_Volterra(x,y,a,b,c,e,t,dt,max_time):
    t_list = []; x_list = []; y_list = []

    # initialize lists
    t_list.append(t); x_list.append(x); y_list.append(y)
    
    while t < max_time:
        # calc new values for t, x, y
        t = t + dt
        x = x + (a*x - b*x*y)*dt
        y = y + (-c*y + e*x*y)*dt
    
        # store new values in lists
        t_list.append(t)
        x_list.append(x)
        y_list.append(y)
    
    # Plot the results    
    p = plt.plot(t_list, x_list, 'b', t_list, y_list, 'r', linewidth = 2)
    
    plt.show()
```

### En primera instancia el código tiene por defecto los siguientes valores  en los siguientes parámetros:
### a = 0.7			 representa la tasa natural de crecimiento de las presas
### b = 0.5			 representa el efecto de la depredación sobre la presa. 
### c = 0.3			 representa la tasa natural de muerte del depredador en ausencia de la presa. 
### e = 0.2			 representa la eficiencia y tasa de propagación del depredador en presencia de la presa. 
### dt = 0.001		 representa la cantidad de tiempo derivada a implementar. 
### max_time = 100	 representa el máximo de tiempo que durara el modelo.
### t = 0			 representa un inicio en el tiempo. Empezando desde 0
### x = 1.0			 representa el numero inicial de presas (cebras).
### y = 0.5			 representa el numero inicial de depredadores(leones).
### Con estos parámetros se puede observar a continuación que la presa cuenta con una alta taza de crecimiento pero a pesar de ello el éxito que tiene el depredador junto con su baja taza de fallecimiento logra reducir drásticamente el numero de presas. Pero conforme mas presas caza este se va quedando cada vez con menos alimento llevando a que el índice de fallecimiento del depredador crezca exponencialmente. 


## MODELO CON VARIABLES POR DEFECTO

```python
a = 0.7; b = 0.5; c = 0.3;  e = 0.2
dt = 0.0001; max_time = 100

# initial time and populations
t = 0; x = 1.0; y = 0.5
funcionLotka_Volterra(x,y,a,b,c,e,t,dt, max_time)
```
