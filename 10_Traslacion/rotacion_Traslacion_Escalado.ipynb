```python
plt.plot([1,1,2,2,1],[1,2,2,1,1])
```

## ROTACION DEL CUADRADO

Se aplica la formula para rotar.

```python
x = [1,1,2,2,1]
y = [1,2,2,1,1]
plt.plot(x, y)

thteta = np.radians(45)
x_rotar = np.cos(thteta) * np.array(x) - np.sin(thteta) * np.array(y)
y_rotar = np.sin(thteta) * np.array(x) + np.cos(thteta) * np.array(y)

plt.plot(x_rotar, y_rotar)

```

## TRASLACION DEL CUADRADO
